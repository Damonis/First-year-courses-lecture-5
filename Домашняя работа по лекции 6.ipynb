2.Сколько в среднем скачиваний у приложений в выборке?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_frame = pd.read_csv('Google-Playstore.csv')
data_frame
```

```python
data_frame['Installs']=data_frame['Installs'].str.replace("+","")
data_frame['Installs']=data_frame['Installs'].str.replace(",","")
average_dowload_quantity=data_frame['Installs'].median()
print("Average dowload is", average_dowload_quantity)
```

Исходя из результата исполнения кода выше можно заключить что среднее число установок приложений = 500

3.Сколько в среднем стоит приложение в выборке?

```python
average_price=data_frame['Price'].mean()
print("Average price is", average_price)
```

Из выведенного результата можно заключить что средняя цена приложений датабазы = 0.10349915833370804

4.Как распределены данные по скачиваниям?

```python
data_frame['Installs'].value_counts().nlargest(100).plot(kind='bar', figsize=(15,5))
plt.title("Распределение данных по скачиваниям")
plt.ylabel('Число приложений')
plt.xlabel('Количество загрузок')
```

Гипстограмма показывает количество приложений в зависимости от их числа загрузок

5.Приложения каких категорий скачивают чаще всего?

```python
most_popular_category=data_frame['Category'].mode().dropna() #мода игнорирует нули
print("Most popular category is", most_popular_category)
```

Данный отрезок кода показал что самая распространненая категория предложений это Education,то есть образование

6.Как распределены данные по рейтингу приложений?

```python
data_frame['Rating'].value_counts().nlargest(50).plot(kind='bar', figsize=(15,5))
plt.title("Распределение данных по рейтингу")
plt.ylabel('Число приложений')
plt.xlabel('Rating')
```

Данная гипстограмма показывает число приложений с определенным рейтингом, где число приложений это числа умноженные на 10^6

7.Как бы вы охарактеризовали самое часто встречающееся значение в наборе данных, какое это прилодение?

```python
dataset_data=["App Name", "App Id","Rating","Rating Count","Installs","Minimum Installs","Maximum Installs","Free","Price","Size","Minimum Android","Developer Id","Developer Website","Developer Email","Released","Privacy Policy","Last Updated","Content Rating","Ad Supported","In App Purchases","Editors Choice"]
most_popular_data=data_frame[dataset_data].mode().dropna() #мода игнорирует нули
print("Most popular data is",most_popular_data)
```

Из кода выше можно понять что самым встречающимся приложением является "Tic Tac Toe" или же крестики-нолики

8.Есть ли зависимость категории приложения от рейтинга приложения?

```python
dependancy = data_frame[['Category', "Rating"]].plot.scatter(x='Rating', y='Category', figsize=(20, 20), marker='$\u25EF$', alpha=0.5)
dependancy.set_xlabel('Rating')
dependancy.set_ylabel('Category')
dependancy.axhline(0, color='grey', lw=1)
print(dependancy.axvline(0, color='grey', lw=1))
```

Выше указанная диаграмма рассеивания демонстрирует зависимость Рейтинга приложения от Категории

9.Как распределены оценки (рейтинги) приложений?


```python
distribution = data_frame[['Rating', "Rating Count"]].plot.scatter(x='Rating', y='Rating Count', figsize=(10, 10), marker='$\u25EF$', alpha=0.5)
distribution.set_xlabel('Rating')
distribution.set_ylabel('Rating count')
distribution.axhline(0, color='grey', lw=1)
print(distribution.axvline(0, color='grey', lw=1))
```

Данная диаграмма показывает распределение рейтинга приложений от количества оценок

10.Как кореллируют между собой рейтинг и количество скачиваний приложения?

```python
data_frame = data_frame.astype({"Installs": str})
cor = data_frame[['Rating',"Installs"]].plot.scatter(x='Rating', y="Installs", figsize=(10, 10), marker='$\u25EF$', alpha=0.5)
cor.set_xlabel('Rating')
cor.set_ylabel('Installs count')
cor.axhline(0, color='grey', lw=1)
print(cor.axvline(0, color='grey', lw=1))
```

Данная диаграмма демонстрирует корреляцию рейтинга и количество скачиваний приложения
